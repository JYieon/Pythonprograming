```python
class Mul:
    def __init__(self, m):
        self.m = m

    def __call__(self, n):   # mul Ìï®ÏàòÎ•º __call__ Ìï®ÏàòÎ™ÖÏúºÎ°ú Î≥ÄÍ≤Ω
        return self.m * n
```

```python
if __name__ == "__main__":
    mul3 = Mul(3)
    mul5 = Mul(5)

    print(mul3(10))  # 30 Ï∂úÎ†•
    print(mul5(10))  # 50 Ï∂úÎ†•
```

```python
def mul(m):           # Ïô∏Î∂Ä Ìï®Ïàò
    def wrapper(n):   # ÎÇ¥Î∂Ä Ìï®Ïàò (ÌÅ¥Î°úÏ†Ä)
        return m * n
    return wrapper
```

```python
if __name__ == "__main__":
    mul3 = mul(3)    # m = 3 Ïù∏ wrapper Ìï®ÏàòÍ∞Ä mul3 Ïóê Ï†ÄÏû•
    mul5 = mul(5)    # m = 5 Ïù∏ wrapper Ìï®ÏàòÍ∞Ä mul5 Ïóê Ï†ÄÏû•

    print(mul3(10))  # m = 3, n = 10 Ïù∏ wrapper Ìï®ÏàòÍ∞Ä Ïã§Ìñâ
    print(mul5(10))  # m = 5, n = 10 Ïù∏ wrapper Ìï®ÏàòÍ∞Ä Ïã§Ìñâ
```

```python
import time

def func1(a, b):
    start = time.time()
    print("Ìï®ÏàòÍ∞Ä Ïã§ÌñâÎê©ÎãàÎã§.")

    val = a + b

    end = time.time()
    print("Ìï®Ïàò ÏàòÌñâÏãúÍ∞Ñ: %f Ï¥à" % (end-start))

    return val
```

```python
if __name__ == "__main__":
    result = func1(1, 2)
    print(result)
```

```python
def func2(a, b):
    start = time.time()
    print("Ìï®ÏàòÍ∞Ä Ïã§ÌñâÎê©ÎãàÎã§.")

    val = a * b

    end = time.time()
    print("Ìï®Ïàò ÏàòÌñâÏãúÍ∞Ñ: %f Ï¥à" % (end-start))

    return val
```

```python
def func1(a, b):
    val = a + b
    return val

def func2(a, b):
    val = a * b
    return val
```

```python
''' Ïã§Ìñâ ÏãúÍ∞Ñ Ï∏°Ï†ï Îç∞ÏΩîÎ†àÏù¥ÌÑ∞ '''
def elapsed(func):                                     # Ìï®ÏàòÎ•º Ïù∏ÌíãÏúºÎ°ú Î∞õÎäîÎã§.
    def wrapper(a, b):
        print('Ìï®ÏàòÍ∞Ä Ïã§ÌñâÎê©ÎãàÎã§.')
        start = time.time()
        result = func(a, b)                            # Ìï®Ïàò Ïã§Ìñâ
        end = time.time()
        print("Ìï®Ïàò ÏàòÌñâÏãúÍ∞Ñ: %f Ï¥à" % (end - start))  # Ìï®Ïàò ÏàòÌñâÏãúÍ∞Ñ
        return result                                  # Ìï®Ïàò Ïã§Ìñâ Í≤∞Í≥º Î∞òÌôò
    return wrapper
```

```python
if __name__ == "__main__":
    deco1 = elapsed(func1)
    result = deco1(1,2)
    print(result)
```

```python
if __name__ == "__main__":
    deco2 = elapsed(func2)
    result = deco2(1,2)
    print(result)
```

```python
@elapsed
def func1(a, b):
    val = a + b
    return val

@elapsed
def func2(a, b):
    val = a * b
    return val
```

```python
if __name__ == "__main__":
    result = func1(1,2)
    print(result)
```

```python
if __name__ == "__main__":
    result = func2(1,2)
    print(result)
```

```python
# ÏöîÏïΩ ÌÖçÏä§Ìä∏ Ï†ÄÏû•
article_summarized = summarize(article, ratio=0.1)
```

```python
import collections
import textwrap
import re
```

```python
# Ï§ÑÎ∞îÍøà Ï†ïÎ†¨
article_align = textwrap.fill(article, width=50)
print(article_align)
```

```python
# Îã®Ïñ¥ Ï∂îÏ∂ú
words = re.findall(r'\w+', article_align)
print(words)
```

```python
# ÎπàÎèÑÏàò ÏÇ∞Ï∂ú
counter = collections.Counter(words)
print(counter)
```

```python
# ÌÇ§ÏõåÎìú Ï∂îÏ∂ú
print(counter.most_common(5))
```

```python
keywords = counter.most_common(5)[1:]
```

```python
print(article_summarized)
```

```python
keys = ['# ' + elem[0] for elem in keywords]
keys = ' '.join(keys)
print(keys)
```
