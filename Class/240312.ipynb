```python
class Mul:
    def __init__(self, m):
        self.m = m

    def mul(self, n):
        return self.m * n
```

```python
if __name__ == "__main__":
    mul3 = Mul(3)
    mul5 = Mul(5)

    print(mul3.mul(10))  # 30 출력
    print(mul5.mul(10))  # 50 출력
```

```python
class Mul:
    def __init__(self, m):
        self.m = m

    def __call__(self, n):   # mul 함수를 __call__ 함수명으로 변경
        return self.m * n
```

```python
if __name__ == "__main__":
    mul3 = Mul(3)
    mul5 = Mul(5)

    print(mul3(10))  # 30 출력
    print(mul5(10))  # 50 출력
```

```python
def mul(m):           # 외부 함수
    def wrapper(n):   # 내부 함수 (클로저)
        return m * n
    return wrapper
```

```python
if __name__ == "__main__":
    mul3 = mul(3)    # m = 3 인 wrapper 함수가 mul3 에 저장
    mul5 = mul(5)    # m = 5 인 wrapper 함수가 mul5 에 저장

    print(mul3(10))  # m = 3, n = 10 인 wrapper 함수가 실행
    print(mul5(10))  # m = 5, n = 10 인 wrapper 함수가 실행
```

```python
import time

def func1(a, b):
    start = time.time()
    print("함수가 실행됩니다.")

    val = a + b

    end = time.time()
    print("함수 수행시간: %f 초" % (end-start))

    return val
```

```python
if __name__ == "__main__":
    result = func1(1, 2)
    print(result)
```

```python
def func2(a, b):
    start = time.time()
    print("함수가 실행됩니다.")

    val = a * b

    end = time.time()
    print("함수 수행시간: %f 초" % (end-start))

    return val
```

```python
def func1(a, b):
    val = a + b
    return val

def func2(a, b):
    val = a * b
    return val
```

```python
''' 실행 시간 측정 데코레이터 '''
def elapsed(func):                                     # 함수를 인풋으로 받는다.
    def wrapper(a, b):
        print('함수가 실행됩니다.')
        start = time.time()
        result = func(a, b)                            # 함수 실행
        end = time.time()
        print("함수 수행시간: %f 초" % (end - start))  # 함수 수행시간
        return result                                  # 함수 실행 결과 반환
    return wrapper
```

```python
if __name__ == "__main__":
    deco1 = elapsed(func1)
    result = deco1(1,2)
    print(result)
```

```python
if __name__ == "__main__":
    deco2 = elapsed(func2)
    result = deco2(1,2)
    print(result)
```

```python
@elapsed
def func1(a, b):
    val = a + b
    return val

@elapsed
def func2(a, b):
    val = a * b
    return val
```

```python
if __name__ == "__main__":
    result = func1(1,2)
    print(result)
```

```python
if __name__ == "__main__":
    result = func2(1,2)
    print(result)
```

> ### **str() 과 repr() 함수**
> - str()과 repr()은 모두 객체를 문자열로 반환하는 함수이다.
> - str() 은 **비공식적**인 문자열을 출력
> - repr() 은 **공식적**인 문자열을 출력 (시스템에서 인식가능)
>
> | 구분        | str()                     | repr()                     |
|-------------|---------------------------|----------------------------|
| 성격        | 비공식적인 문자열을 출력  | 공식적인 문자열을 출력     |
| 사용 목적   | 사용자가 보기 쉽도록      | 문자열로 객체를 다시 생성  |
| 누구를 위해 | 프로그램 사용자(end user) | 프로그램 개발자(developer) |

```python
regex = r'0\d{1,2}[ -]?\d{3,4}[ -]?\d{3,4}'
```

| 파일열기모드 | 설명                                                       |
|--------------|------------------------------------------------------------|
| r            | 읽기모드 - 파일을 읽기만 할 때 사용                        |
| w            | 쓰기모드 - 파일에 내용을 쓸 때 사용                        |
| a            | 추가모드 - 파일의 마지막에 새로운 내용을 추가 시킬 때 사용 |

```python
# 텍스트 파일 생성
f = open("새파일.txt", 'w')
f.close()
```

```python
# 텍스트 파일 쓰기
f = open("새파일.txt", 'w')
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
# 한줄 읽기
f = open("새파일.txt", 'r')
line = f.readline()
print(line)
f.close()
```

```python
# 여러줄 읽기
f = open("새파일.txt", 'r')
line = f.readline()
print(line)
line = f.readline()
print(line)
line = f.readline()
print(line)
f.close()
```

```python
# 모든 줄 읽기
f = open("새파일.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
# w 모드 사용 시
f = open("새파일.txt",'w')
for i in range(11, 21):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
# a 모드 사용 시
f = open("새파일.txt",'a')
for i in range(11, 21):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
# close() 사용 불필요
with open("새파일.txt", "w") as f:
    for i in range(1, 11):
        data = "%d번째 줄입니다.\n" % i
        f.write(data)
```

```python
# 요약 텍스트 저장
article_summarized = summarize(article, ratio=0.1)
```

```python
import collections
import textwrap
import re
```

```python
# 줄바꿈 정렬
article_align = textwrap.fill(article, width=50)
print(article_align)
```

```python
# 단어 추출
words = re.findall(r'\w+', article_align)
print(words)
```

```python
# 빈도수 산출
counter = collections.Counter(words)
print(counter)
```

```python
# 키워드 추출
print(counter.most_common(5))
```

```python
keywords = counter.most_common(5)[1:]
```

```python
print(article_summarized)
```

```python
keys = ['# ' + elem[0] for elem in keywords]
keys = ' '.join(keys)
print(keys)
```
